В домашней работе вам необходимо, используя пример из урока 19.2 и Keras 3, произвести тонкую настройку модели второй версии [EfficientNetV2](https://keras.io/api/applications/efficientnet_v2/) на примере набора данных [Stanford Dogs](https://storage.yandexcloud.net/academy.ai/stanford_dogs.zip).

```python
import matplotlib.pyplot as plt
%matplotlib inline

from tensorflow import data as tf_data
import os
import keras
from keras import layers
from keras.applications import EfficientNetV2B0
import tensorflow as tf
import numpy as np
from sklearn.utils.class_weight import compute_class_weight

# Установим seed для воспроизводимости
tf.random.set_seed(42)
np.random.seed(42)

IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32

# Загрузка данных
!wget -q https://storage.yandexcloud.net/academy.ai/stanford_dogs.zip
!unzip -qo "stanford_dogs" -d ./dataset

IMAGE_PATH = './dataset/'

# Удаление поврежденных изображений
num_skipped = 0
for folder_name in os.listdir(IMAGE_PATH):
    folder_path = os.path.join(IMAGE_PATH, folder_name)
    for fname in os.listdir(folder_path):
        fpath = os.path.join(folder_path, fname)
        try:
            fobj = open(fpath, "rb")
            is_jfif = b"JFIF" in fobj.peek(10)
        finally:
            fobj.close()
        if not is_jfif:
            num_skipped += 1
            os.remove(fpath)

print(f"Удалено изображений: {num_skipped}")

# Получаем список всех классов
all_classes = sorted(os.listdir(IMAGE_PATH))
print(f"Всего классов: {len(all_classes)}")

# Выбираем случайные 10 пород
selected_classes = sorted(np.random.choice(all_classes, 10, replace=False))
print(f"Выбранные 10 пород: {selected_classes}")

# Создаем временную директорию для выбранных классов
!rm -rf ./selected_dataset
!mkdir -p ./selected_dataset

for class_name in selected_classes:
    !cp -r "{os.path.join(IMAGE_PATH, class_name)}" "./selected_dataset/"

SELECTED_IMAGE_PATH = './selected_dataset/'

# Разделение данных
train_ds, val_ds = keras.utils.image_dataset_from_directory(
    SELECTED_IMAGE_PATH,
    validation_split=0.2,
    subset="both",
    seed=42,
    shuffle=True,
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
)

# Создаем тестовую выборку из валидационной
val_batches = tf.data.experimental.cardinality(val_ds)
test_ds = val_ds.take(val_batches // 2)
val_ds = val_ds.skip(val_batches // 2)

CLASS_LIST = sorted(os.listdir(SELECTED_IMAGE_PATH))
CLASS_COUNT = len(CLASS_LIST)

print(f'Количество классов: {CLASS_COUNT}')
print(f'Метки классов: {CLASS_LIST}')

# Визуализация примеров
plt.figure(figsize=(12, 12))
for images, labels in train_ds.take(1):
    for i in range(min(9, len(images))):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(keras.utils.img_to_array(images[i]).astype("uint8"))
        plt.title(CLASS_LIST[labels[i]])
        plt.axis("off")
plt.show()

# аугментация данных
def create_augmentation_model():
    inputs = tf.keras.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3))
    x = layers.RandomFlip("horizontal")(inputs)
    x = layers.RandomRotation(0.2)(x)  # Увеличили rotation
    x = layers.RandomZoom(0.2)(x)      # Увеличили zoom
    x = layers.RandomContrast(0.2)(x)  # Увеличили contrast
    x = layers.RandomBrightness(0.2)(x) # Добавили brightness
    x = layers.RandomTranslation(0.1, 0.1)(x) # Добавили translation
    # Дополнительные аугментации
    x = layers.RandomCrop(IMAGE_SIZE[0], IMAGE_SIZE[1])(x)  # Случайный кроп
    x = layers.GaussianNoise(0.01)(x)  # Добавили шум
    return tf.keras.Model(inputs, x)

augmentation_model = create_augmentation_model()

# Функции для обработки данных
def normalize_img(image, label):
    # Правильная нормализация для EfficientNet
    image = tf.cast(image, tf.float32)
    image = tf.keras.applications.efficientnet.preprocess_input(image)
    return image, label

def apply_augmentation(images, labels):
    augmented_images = augmentation_model(images, training=True)
    return augmented_images, labels

# Подготовка данных
train_ds = train_ds.map(normalize_img, num_parallel_calls=tf_data.AUTOTUNE)
val_ds = val_ds.map(normalize_img, num_parallel_calls=tf_data.AUTOTUNE)
test_ds = test_ds.map(normalize_img, num_parallel_calls=tf_data.AUTOTUNE)

# Применяем аугментацию к обучающей выборке
train_ds = train_ds.map(apply_augmentation, num_parallel_calls=tf_data.AUTOTUNE)

# One-hot кодирование
train_ds = train_ds.map(
    lambda img, label: (img, tf.one_hot(label, CLASS_COUNT)),
    num_parallel_calls=tf_data.AUTOTUNE
)
val_ds = val_ds.map(
    lambda img, label: (img, tf.one_hot(label, CLASS_COUNT)),
    num_parallel_calls=tf_data.AUTOTUNE
)
test_ds = test_ds.map(
    lambda img, label: (img, tf.one_hot(label, CLASS_COUNT)),
    num_parallel_calls=tf_data.AUTOTUNE
)

# Вычисляем веса классов для балансировки
def get_class_weights(dataset):
    labels = []
    for _, batch_labels in dataset:
        labels.extend(tf.argmax(batch_labels, axis=1).numpy())

    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=np.unique(labels),
        y=labels
    )
    return dict(enumerate(class_weights))

class_weights = get_class_weights(train_ds)
print("Веса классов для балансировки:", class_weights)

# Кэширование и предварительная выборка
train_ds = train_ds.cache().prefetch(tf_data.AUTOTUNE)
val_ds = val_ds.cache().prefetch(tf_data.AUTOTUNE)
test_ds = test_ds.cache().prefetch(tf_data.AUTOTUNE)

#  построение модели с регуляризацией
def build_model(num_classes):
    # Базовая модель
    base_model = EfficientNetV2B0(
        include_top=False,
        weights="imagenet",
        input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3),
        pooling='avg'
    )

    # Заморозка базовой модели
    base_model.trainable = False

    # Создаем полную модель
    inputs = keras.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3))
    x = base_model(inputs, training=False)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.6)(x)

    # Дополнительные слои
    x = layers.Dense(512, activation='relu',
                    kernel_regularizer=keras.regularizers.l2(0.001))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)

    # Добавим еще один слой
    x = layers.Dense(256, activation='relu',
                    kernel_regularizer=keras.regularizers.l2(0.001))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(num_classes, activation="softmax",
                          kernel_regularizer=keras.regularizers.l2(0.001))(x)

    model = keras.Model(inputs, outputs)

    # Компиляция с оптимизированными параметрами
    optimizer = keras.optimizers.Adam(
        learning_rate=0.0001,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-07,
        amsgrad=False
    )

    model.compile(
        optimizer=optimizer,
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )

    return model, base_model

# Создаем модель
model, base_model = build_model(CLASS_COUNT)
model.summary()

# Callbacks
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=12,
        restore_best_weights=True,
        verbose=1,
        min_delta=0.001,
        mode='max'
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=6,
        min_lr=1e-7,
        verbose=1,
        mode='max'
    ),
    keras.callbacks.ModelCheckpoint(
        filepath='best_model_effnetv2.keras',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    ),
    # Добавим Learning Rate Scheduler
    keras.callbacks.LearningRateScheduler(
        lambda epoch: 0.0001 * (0.9 ** epoch) if epoch > 10 else 0.0001
    )
]

# Первый этап обучения
print("Первый этап обучения (замороженные веса)...")
history = model.fit(
    train_ds,
    epochs=35,
    validation_data=val_ds,
    callbacks=callbacks,
    verbose=1,
    class_weight=class_weights
)

# Визуализация
def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    ax1.plot(history.history['accuracy'], label='Обучающая точность')
    ax1.plot(history.history['val_accuracy'], label='Валидационная точность')
    ax1.set_title('Точность модели')
    ax1.set_xlabel('Эпоха')
    ax1.set_ylabel('Точность')
    ax1.legend()
    ax1.grid(True)

    ax2.plot(history.history['loss'], label='Обучающие потери')
    ax2.plot(history.history['val_loss'], label='Валидационные потери')
    ax2.set_title('Потери модели')
    ax2.set_xlabel('Эпоха')
    ax2.set_ylabel('Потери')
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

plot_training_history(history)

print("\nТонкая настройка...")

# Размораживаем базовую модель
base_model.trainable = True

# Заморозка  первых 100 слоев
for layer in base_model.layers[:100]:
    if not isinstance(layer, layers.BatchNormalization):
        layer.trainable = False

print(f"Обучаемых слоев в базовой модели: {sum([layer.trainable for layer in base_model.layers])}/{len(base_model.layers)}")

# Перекомпиляция с очень маленьким learning rate
optimizer = keras.optimizers.Adam(
    learning_rate=0.00001,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07
)
model.compile(
    optimizer=optimizer,
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# Тонкая настройка
print("Начинаем тонкую настройку...")
history_fine = model.fit(
    train_ds,
    epochs=25,
    initial_epoch=len(history.history['accuracy']),
    validation_data=val_ds,
    callbacks=callbacks,
    verbose=1,
    class_weight=class_weights
)

# Объединенная история
full_history = {
    'accuracy': history.history['accuracy'] + history_fine.history['accuracy'],
    'val_accuracy': history.history['val_accuracy'] + history_fine.history['val_accuracy'],
    'loss': history.history['loss'] + history_fine.history['loss'],
    'val_loss': history.history['val_loss'] + history_fine.history['val_loss']
}

plot_training_history(type('History', (), {'history': full_history}))

# Test-Time Augmentation для улучшения точности
def predict_with_tta(model, dataset, n_augmentations=7):
    """Предсказание с Test-Time Augmentation"""
    all_predictions = []
    all_labels = []

    for images, labels in dataset:
        augmented_predictions = []
        for _ in range(n_augmentations):
            augmented_images = augmentation_model(images, training=True)
            preds = model(augmented_images, training=False)
            augmented_predictions.append(preds)

        # Усредняем предсказания
        avg_preds = tf.reduce_mean(augmented_predictions, axis=0)
        all_predictions.append(avg_preds)
        all_labels.append(labels)

    return tf.concat(all_predictions, axis=0), tf.concat(all_labels, axis=0)

print("Применяем Test-Time Augmentation для улучшения точности...")

# Оценка с TTA
tta_val_predictions, tta_val_labels = predict_with_tta(model, val_ds)
tta_test_predictions, tta_test_labels = predict_with_tta(model, test_ds)

# Вычисляем точность с TTA
tta_val_accuracy = tf.reduce_mean(
    tf.cast(tf.argmax(tta_val_predictions, axis=1) == tf.argmax(tta_val_labels, axis=1), tf.float32)
)
tta_test_accuracy = tf.reduce_mean(
    tf.cast(tf.argmax(tta_test_predictions, axis=1) == tf.argmax(tta_test_labels, axis=1), tf.float32)
)

# Обычная оценка
print("\nОценка модели:")
train_loss, train_acc = model.evaluate(train_ds, verbose=0)
val_loss, val_acc = model.evaluate(val_ds, verbose=0)
test_loss, test_acc = model.evaluate(test_ds, verbose=0)

print(f"Обучающая выборка - Точность: {train_acc:.4f}, Потери: {train_loss:.4f}")
print(f"Валидационная выборка - Точность: {val_acc:.4f}, Потери: {val_loss:.4f}")
print(f"Тестовая выборка - Точность: {test_acc:.4f}, Потери: {test_loss:.4f}")

print(f"\nС Test-Time Augmentation:")
print(f"Валидационная выборка TTA - Точность: {tta_val_accuracy:.4f}")
print(f"Тестовая выборка TTA - Точность: {tta_test_accuracy:.4f}")

# Проверка условий
print("ФИНАЛЬНЫЕ РЕЗУЛЬТАТЫ:")
print(f"Точность на обучающей выборке: {train_acc:.2%}")
print(f"Точность на валидационной выборке: {val_acc:.2%}")
print(f"Точность на тестовой выборке: {test_acc:.2%}")
print(f"Точность на валидационной выборке с TTA: {tta_val_accuracy:.2%}")
print(f"Точность на тестовой выборке с TTA: {tta_test_accuracy:.2%}")

# Используем лучшую точность (обычную или TTA)
final_val_acc = max(val_acc, tta_val_accuracy.numpy())
final_test_acc = max(test_acc, tta_test_accuracy.numpy())

print(f"\nЛучшая точность на валидации: {final_val_acc:.2%}")
print(f"Лучшая точность на тесте: {final_test_acc:.2%}")

# Сохранение модели
model.save('final_effnetv2_model.keras')
print("Модель сохранена!")
```
